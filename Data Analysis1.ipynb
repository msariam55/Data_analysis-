```python
x = DF.groupby(['Year']).mean()
y = x['TotalPay']
fig, ax = plt.subplots(figsize=(8, 5),facecolor='lightblue')
ax.bar(x.index, y, label='Barplot1')
ax.set_xlabel('Year')
ax.set_ylabel('Mean TotalPay')
ax.set_title('Relation Between year and totalPay')
plt.legend()
plt.show()
```

```python
y=DF['JobTitle'].head(200)
x= DF['JobTitle'].value_counts().head(200)
fig, ax = plt.subplots(figsize=(10,12))
plt.scatter(x,y,label='plot')
y_tick_lines = ax.yaxis.get_ticklines()
for line in y_tick_lines:
    line.set_linewidth(0.0001)
plt.grid( color='green', linestyle='--')
plt.legend()
plt.show()
```

```python
y=DF['BasePay'].value_counts().head(30)
y2=DF['OvertimePay'].value_counts().head(30)
x=DF['JobTitle'].head(30)
fig, ax = plt.subplots(figsize=(6, 6))

plt.plot(x,y,'r',label='BasePay')
plt.plot(x, y2, 'g', label='OverTimePay')
ax.set_title('BasePay&OverTimePay')
ax.set_ylim(0,1000)
ax.set_xticklabels(x, rotation='vertical')
plt.legend()
plt.show()
```

```python
DF.replace('Not Provided', np.nan, inplace=True)
DF['BasePay'] = pd.to_numeric(DF['BasePay'])
```

Salary trend by year

```python
DF.groupby('Year').agg({'TotalPay': 'mean'}).plot(kind='line')
plt.show()
```

Salary trend by job


```python
DF.groupby('JobTitle').agg({'TotalPay': 'mean'}).sort_values(by='TotalPay', ascending=False).head(10).plot(kind='barh')
plt.show()
```
